```python
import numpy as np
```

```python
import jsonlines
datas = []
with jsonlines.open("rank_evaluate/promptway_own|targetlm_do_sample_False|append_label_length_-1.jsonl") as f:
    for line in f:
        datas.append(line)

```

```python
ranks = []
for data in datas:
    ranks.append(data["rank"])
```

```python
max_length = max(len(l) for l in ranks)
padded_lists = [np.pad(l, (0, max_length - len(l)), 'constant') for l in ranks]
stacked_array = np.stack(padded_lists)
```

```python
stacked_array.shape
```

```python
stacked_array
```

```python
import plotly.express as px
```

```python
summed_array = np.sum(stacked_array, axis=0)
fig = px.bar(x=np.arange(len(summed_array)), y=summed_array, labels={'x': 'Index', 'y': 'Sum'})
```

```python
fig.show()
```
