```python
from collections import defaultdict as ddict
case_step__loss_d = {'case1': ddict(list), 'case2': ddict(list)}

# Classify and count cases by step
for item in data_list:
    step = int(item['step'].replace("step_",""))
    harm_score = item['harm_score']
    harm_em = item['harm_em']
    loss = item["loss"]
    
    if harm_score == 1 and harm_em == 1:
        case_key = 'case2'
    else:
        case_key = 'case1'
    
    # Round the step to the nearest interval of 50 for aggregation
    rounded_step = 50 * ((step + 1) // 50)
    case_step__loss_d[case_key][rounded_step].append(loss)
```

```python
from collections import defaultdict as ddict
case_step__loss_d = {'case1': ddict(list), 'case2': ddict(list), 'case3': ddict(list)}

# Classify and count cases by step
for item in data_list:
    step = int(item['step'].replace("step_",""))
    harm_score = item['harm_score']
    harm_em = item['harm_em']
    loss = item["loss"]
    
    if harm_score == 1 and harm_em == 1:
        case_key = 'case2'
    elif
    else:
        case_key = 'case1'
    
    # Round the step to the nearest interval of 50 for aggregation
    rounded_step = 50 * ((step + 1) // 50)
    case_step__loss_d[case_key][rounded_step].append(loss)
```

```python
import random
random.seed(42)
def proportional_sample(len_case1,len_case2,count_samples_from_case2 = 100):
    count_samples_from_case2 = min(count_samples_from_case2,len_case2)
    count_samples_from_case1 = min(int(count_samples_from_case2 * len_case1/len_case2),len_case1)
    return count_samples_from_case1,count_samples_from_case2

case_step__loss_d_for_plot = {'case1': ddict(list), 'case2': ddict(list)}
for key in case_step__loss_d['case1'].keys():
    count_samples_from_case1,count_samples_from_case2 = proportional_sample(len(case_step__loss_d['case1'][key]),len(case_step__loss_d['case2'][key]))
    samples_from_case1,samples_from_case2 = random.sample(case_step__loss_d['case1'][key],count_samples_from_case1),random.sample(case_step__loss_d['case2'][key],count_samples_from_case2)
    samples_from_case1.append(min(case_step__loss_d['case1'][key]))
    samples_from_case2.append(min(case_step__loss_d['case2'][key]))
    case_step__loss_d_for_plot["case1"][key] = samples_from_case1
    case_step__loss_d_for_plot["case2"][key] = samples_from_case2
```

```python

case_step__loss_d_for_plot_processed = {'case1': ddict(list), 'case2': ddict(list)}
for key1 in case_step__loss_d_for_plot:
    for key2 in case_step__loss_d_for_plot[key1]:
        case_step__loss_d_for_plot_processed[key1][key2] = np.log(case_step__loss_d_for_plot[key1][key2])
import numpy as np
import plotly.graph_objects as go

fig = go.Figure()

# ‰∏∫ÊØè‰∏™caseÂÆö‰πâ‰∏ÄËá¥ÁöÑÈ¢úËâ≤
colors = {'case1': 'blue', 'case2': 'red'}

# ‰øùËØÅÂõæ‰æã‰∏≠Âè™ÊòæÁ§∫‰∏ÄÊ¨°ÊØè‰∏™case
legend_added = {'case1': False, 'case2': False}
legend_name = {'case1': "Safe Points", 'case2': "Harmful Points"}

# ÂàùÂßãÂåñËÆ∞ÂΩïÊúÄ‰ΩéÁÇπÁöÑÂàóË°®
lowest_loss_points = []
min_loss_so_far = np.inf

# È¶ñÂÖàÁ°ÆÂÆöÊâÄÊúâÊ≠•È™§
all_steps = sorted(set(step for steps in case_step__loss_d_for_plot_processed.values() for step in steps))

# ÈÅçÂéÜÊØè‰∏™Ê≠•È™§ÔºåÊâæÂá∫ËØ•Ê≠•È™§ÊâÄÊúâcaseÁöÑÊúÄ‰ΩélossÔºåÂπ∂ËÆ∞ÂΩïÊâÄÊúâÁÇπ
for step in all_steps:
    current_step_min_loss = np.inf  # ÂàùÂßãÂåñÂΩìÂâçÊ≠•È™§ÁöÑÊúÄ‰Ωéloss
    for case, steps in case_step__loss_d_for_plot_processed.items():
        if step in steps:
            losses = steps[step]
            # Êõ¥Êñ∞ÂΩìÂâçÊ≠•È™§ÁöÑÊúÄ‰Ωéloss
            current_step_min_loss = min(current_step_min_loss, min(losses))

            # ÁîªÂá∫ÂΩìÂâçcaseÁöÑÊâÄÊúâÁÇπ
            fig.add_trace(go.Scatter(
                x=[step] * len(losses),
                y=losses,
                mode='markers',
                name=legend_name[case] if not legend_added[case] else None,
                marker=dict(size=10, color=colors[case]),
                legendgroup=case,
                showlegend=not legend_added[case]
            ))
            legend_added[case] = True

    # Ê£ÄÊü•ÂΩìÂâçÊ≠•È™§ÁöÑÊúÄ‰ΩélossÊòØÂê¶‰∏∫ÂÖ®Â±ÄÊúÄ‰Ωé
    if current_step_min_loss < min_loss_so_far:
        min_loss_so_far = current_step_min_loss
    lowest_loss_points.append((step, min_loss_so_far))

# Ê†πÊçÆËÆ∞ÂΩïÁöÑÊúÄ‰ΩéÊçüÂ§±ÁÇπËøûÁ∫ø
if lowest_loss_points:
    steps, losses = zip(*lowest_loss_points)
    fig.add_trace(go.Scatter(
        x=steps,
        y=losses,
        mode='lines+markers',
        name='Anchor Point for Next Steps',
        marker=dict(size=10, color='green'),
        line=dict(color='green')
    ))


fig.update_layout(
    title='Aggregated Visualization of Ten cases from GCG optimization',
    xaxis_title='Steps',
    yaxis_title='Log(Loss)',
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=1,
        title_font=dict(size=16, color='black', family='Arial, sans-serif'),  # ËÆæÁΩÆÂõæ‰æãÊ†áÈ¢òÁöÑÂ≠ó‰Ωì
        font=dict(size=14, color='black', family='Arial, sans-serif')  # ËÆæÁΩÆÂõæ‰æãÈ°πÁöÑÂ≠ó‰Ωì
    ),
    height=800,
    xaxis=dict(
        title='Steps',
        title_font=dict(size=18, color='black', family='Arial, sans-serif'),  # ËÆæÁΩÆXËΩ¥Ê†áÈ¢òÁöÑÂ≠ó‰Ωì
    ),
    yaxis=dict(
        title='Log(Loss)',
        title_font=dict(size=18, color='black', family='Arial, sans-serif'),  # ËÆæÁΩÆYËΩ¥Ê†áÈ¢òÁöÑÂ≠ó‰Ωì
    )
)


fig.show()
fig.write_image(f"./figures/{model}_optimization_process.pdf",width=1200, height=800)
```

```python
import glob
import os
import json
import re
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def find_files(pattern):
    # ‰ΩøÁî®glob.globÊêúÁ¥¢ÂåπÈÖçÁâπÂÆöÊ®°ÂºèÁöÑÊñá‰ª∂
    return glob.glob(pattern,recursive= True)

def extract_number_from_string(s):
    # ÂÆö‰πâÊ≠£ÂàôË°®ËææÂºèÊ®°Âºè
    pattern = re.compile(r'checkpoint-(\d+)-')

    # Âú®Â≠óÁ¨¶‰∏≤‰∏≠ÊêúÁ¥¢Ê®°Âºè
    match = pattern.search(s)
    # Â¶ÇÊûúÊâæÂà∞ÂåπÈÖçÈ°πÔºåÂàôËøîÂõûÊèêÂèñÁöÑÊï∞Â≠óÔºåÂê¶ÂàôËøîÂõûNone
    if match:
        return match.group(1)
    else:
        return None

def extract_decode_from_string(s):
    # ÂÆö‰πâÊ≠£ÂàôË°®ËææÂºèÊ®°Âºè
    pattern = re.compile(r'\|decode_(\w+)\|')

    # Âú®Â≠óÁ¨¶‰∏≤‰∏≠ÊêúÁ¥¢Ê®°Âºè
    match = pattern.search(s)
    # Â¶ÇÊûúÊâæÂà∞ÂåπÈÖçÈ°πÔºåÂàôËøîÂõûÊèêÂèñÁöÑÊï∞Â≠óÔºåÂê¶ÂàôËøîÂõûNone
    if match:
        return match.group(1)
    else:
        return None


from collections import defaultdict as ddict
def convert_to_d(files):
    d = ddict(dict)
    for file in files:
        with open(file) as f:
            data = json.load(f)
        ckpt_steps = extract_number_from_string(file)
        ckpt_decode = extract_decode_from_string(file)
        # print(ckpt_steps,'ckpt_steps')
        d[ckpt_steps][ckpt_decode]=data
    return d


import pathlib
def plot(data, sample_way = None, split = None,target_lm = None,save = False,path = None):
    if save:
        assert path is not None
        pathlib.Path(path).parent.mkdir(exist_ok= True, parents= True)
        
    # ÂàõÂª∫ÂÖ∑Êúâ‰∏§‰∏™yËΩ¥ÁöÑÂõæÂΩ¢
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    # ÂáÜÂ§áÂ≠òÂÇ®ÊØèÁßç decode Á±ªÂûãÁöÑÊï∞ÊçÆ
    decode_data = {}

    # Ëø≠‰ª£ÊØè‰∏™Ê≠•È™§Êù•Êî∂ÈõÜÊï∞ÊçÆ
    for step in sorted(data.keys(), key=lambda x: int(x)):
        for decode, values in data[step].items():
            if decode not in decode_data:
                decode_data[decode] = {'steps': [], 'jb_values': []}
            decode_data[decode]['steps'].append(step)
            decode_data[decode]['jb_values'].append(values['asr_over_all_qs'])

    # ‰∏∫ÊØèÁßç decode Á±ªÂûãÊ∑ªÂä†‰∏ÄÊù°Á∫ø
    for decode, values in decode_data.items():
        fig.add_trace(
            go.Scatter(
                x=values['steps'], y=values['jb_values'], mode='lines+markers',
                name=f'{decode.replace("_numreturn_","+")}'
            ),
            secondary_y=False,
        )

    # ËÆæÁΩÆyËΩ¥ÁöÑËåÉÂõ¥
    fig.update_yaxes(title_text="ASR(%)", range=[0, 1], secondary_y=False)

    # Êõ¥Êñ∞Â∏ÉÂ±Ä
    fig.update_layout(
        title_font_size=14,
        xaxis_title="Steps",
        legend_title="Sampling",
        width=800
    )

    # ÊòæÁ§∫ÂõæË°®
    if not save:
        fig.show()
    elif save:
        fig.write_image(path)
```
